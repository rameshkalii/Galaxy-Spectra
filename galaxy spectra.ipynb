```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
file = fits.open('galaxy.fits')
# file.info()
data= file[1].data
flux = data['flux']
wav = 10**data['loglam']
plt.plot(wav,flux)   #Plotting
plt.title('Galaxy Spectra of ARP 106')
plt.xlabel(r'Wavelength $ \AA$')
plt.ylabel(r'Flux $ 10^{-17} erg/cm^2/s/\AA$')

# Redshift
z = file[2].data['z']
z = z[0]
z
```

```python
## Spectra for rest frame
```

```python
plt.figure()
plt.plot(wav/(1+z),flux,color='red',label='Spectra')
plt.title('Galaxy Spectra ARP 106 in Rest frame')
plt.xlabel(r'Wavelength $ \AA$')
plt.ylabel(r'flux $ 10^{-17} erg/cm^2/s\AA$')
plt.annotate("H-alpha",
           xy = (6564.16,188),
           xytext = (6564.46,188.3))
plt.annotate("O-II",
           xy = (3727.092,137),
           xytext = (3727.392,137.3))
plt.annotate("O-I",
           xy = (6302.046,130),
           xytext = (6302.346,130.3))
plt.savefig('Galaxy Spectra of ARP 106')
```
